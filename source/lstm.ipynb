```python
# csv 파일 경로
csv_path = '/home/siwon/dev/Deeplearning-6/data/trajectory_dataset.csv'

# csv 파일 읽기(no column name, 세로로 불러오기)
df = pd.read_csv(csv_path, header=None, sep=';')
df
```

```python
def add_index_to_elements(df):
    new_data = []
    for idx, row in df.iterrows():
        new_row = []
        for col, elem_str in enumerate(row):
            # 문자열로 저장된 리스트들을 파이썬 리스트로 변환
            elem_str = elem_str.replace('"', '').strip('][')
            elem_lists = elem_str.split('],[')
            
            new_elem_lists = []
            for elem_list in elem_lists:
                # 불필요한 쉼표 처리
                if not elem_list or elem_list == ',':
                    continue
                
                elem_list = elem_list.strip('],').split(', ')
                elem = [int(x) for x in elem_list if x != ''] + [col]
                new_elem_lists.append(elem)
            
            new_row.extend(new_elem_lists)
        new_data.append(new_row)
    return new_data

updated_data = add_index_to_elements(df)
```

```python
import pandas as pd
import re

def parse_list_string(list_string):
    list_string = list_string.strip('][')
    list_string = re.sub(r'\s+', '', list_string)
    elements = list_string.split(',')
    return [int(x) for x in elements if x != '']

def preprocess_dataframe(df):
    for col in df.columns:
        df[col] = df[col].apply(parse_list_string)
    return df

updated_df = preprocess_dataframe(df)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
print(updated_df)
```
